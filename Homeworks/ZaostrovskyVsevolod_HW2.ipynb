### Problem 1 (40 points)

Let $S^{(r)}$ be the underlying process from the CEV model with interest rate $r$, that is satisfying the SDE

$$dS^{(r)}_t = r S^{(r)}_t dt + \sigma (S^{(r)}_t)^{\beta} dW_t, \quad S^{(r)}_0 = s > 0.$$

Show that under the time-change 
$$\tau(t) = \frac{e^{2r(\beta - 1)t} - 1}{2r(\beta - 1)},$$

the following relationship holds 

$$S^{(r)}_t \stackrel{d}{=} e^{rt} S^{(0)}_{\tau(t)}.$$

*Instructions:*

**1**. ðŸ§  <span style="color:blue">(10 points)</span>  Show that the discounted underlying process process $X_t := e^{-rt} S_t^{(r)}$ satisfies the SDE

$$dX_t = \sigma e^{r(\beta - 1)t} X_t^{\beta} dW_t.$$

<span style="color:red">Solution:</span>

$$ d X_t = e^{-r t} d S_t^{(r)} + S_t^{(r)} d e^{-r t} + d S_t^{(r)} d e^{-r t} = e^{-r t} d S_t^{(r)} - r S_t^{(r)} e^{-r t} d t = e^{-r t} (r S^{(r)}_t dt + \sigma (S^{(r)}_t)^{\beta} dW_t) - r S_t^{(r)} e^{-r t} d t =  e^{-r t} \sigma (S^{(r)}_t)^{\beta} dW_t, $$ 

$$ \sigma (S^{(r)}_t)^{\beta} dW_t = \sigma X_t ^{\beta} e^{r t \beta}, $$

$$ d X_t = e^{-r t} \sigma (S^{(r)}_t)^{\beta} dW_t = e^{-r t} \sigma X_t ^{\beta} e^{r t \beta} dW_t = e^{r t (\beta - 1)} \sigma X_t ^{\beta} dW_t.$$

**2**. ðŸ§  <span style="color:blue">(10 points)</span>  Show that the time-changed zero-rate underlying process process $\widetilde X_t := S_{\tau(t)}^{(0)}$ can be represented as

$$\widetilde X_t = s_0 + \int_{0}^{t} \sigma (S^{(0)}_{\tau(u)})^{\beta}dW_{\tau(u)}.$$


Then show that $W_{\tau(u)}$ can be represented as 
$$W_{\tau(u)} = \int_{0}^{u} \sqrt{\tau'(v)} d \widetilde W_v,$$

where $\tilde W_v$ is some Brownian motion. 

<span style="color:red">Solution:</span>

$$\widetilde X_t = S_{\tau(t)}^{(0)} = s_0 + \int\limits_{0}^{\tau(t)} \sigma (S^{(0)}_{u})^{\beta}dW_{u} =  s_0 + \int\limits_{0}^{t} \sigma (S^{(0)}_{\tau(u)})^{\beta}dW_{\tau(u)}.$$

$$ E W_{\tau(u)} = E (\int_{0}^{u} \sqrt{\tau'(v)} d \widetilde {W_v}) = 0. $$

$$ cov(W_{\tau(a)}, W_{\tau(b)}) = \min (\tau(a), \tau(b)). $$

$$ cov(\int_{0}^{a} \sqrt{\tau'(v)} d \widetilde {W_v}, \int_{0}^{b} \sqrt{\tau'(v)} d \widetilde {W_v}) = E \int_{0}^{a} \sqrt{\tau'(v)} d \widetilde {W_v} \int_{0}^{b} \sqrt{\tau'(v)} d \widetilde {W_v} = E (\int_{0}^{\min (a, b)} \sqrt{\tau'(v)} d \widetilde {W_v})^2 =\\
= (\text{Ito's isometry}) =E \int\limits_{0}^{\min(m,\ n)}\tau'(v) dv = \min(\tau(m), \tau(n)). $$

Thus, we have: $W_{\tau(u)} = \int_{0}^{u} \sqrt{\tau'(v)} d \widetilde W_v .$


**3**. ðŸ§  <span style="color:blue">(10 points)</span>  Show that $\widetilde X_t$ satisfies the SDE

<span style="color:red">Solution:</span>
$$ \tau'(t) = \frac{2r(\beta - 1)e^{2r(\beta-1)t}}{2r(\beta - 1)} = e^{2r(\beta-1)t}. $$
$$ d \widetilde X_t = \sigma (S^{(0)}_{\tau(t)})^{\beta}dW_{\tau(t)} = \sigma (S^{(0)}_{\tau(t)})^{\beta} \sqrt{\tau'(t)} d \widetilde W_t =
e^{r(\beta-1)t} \sigma (\widetilde {X}_t)^{\beta} d \widetilde{W_t}.$$
